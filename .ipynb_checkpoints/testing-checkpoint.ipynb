```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
```

```python
img = mpimg.imread('lakes/leech_lake.png')
imgplot = plt.imshow(img)
```

```python
pixels = 0
for row in img:
    for pixel in row:
        pixels += 1

print(f'There are {pixels} pixels in the image')
```

```python
print( img[0][0] )
```

```python
bw_img = []
for row in img:
    bw_row = []
    for pixel in row:
        if pixel[1] > pixel[2]:
            bw_row.append(0)
        else:
            bw_row.append(1)
    bw_img.append(bw_row)
    
```

```python
plt.imshow(bw_img)
```
